## Trabalhando com Excel no Pandas

```python
# Importanto biblioteca
import pandas as pd
```

```python
# Importando arquivos excel
df1 = pd.read_excel("/content/drive/MyDrive/datasets_DIO/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/datasets_DIO/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/datasets_DIO/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/datasets_DIO/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/datasets_DIO/Salvador.xlsx")
```

```python
#Concatenando em um DataFrame √∫nico
df = pd.concat([df1,df2,df3,df4,df5])
df.sample(5)
```

```python
# Verificando o tipo de dados
df.dtypes
```

```python
# Alterando o tipo de dado de LojaID
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

```python
# Tratando valores Nulos:


# Consultando se h√° valores nulos:
df.isnull().sum()
```

```python
# Caso tivesse (por exemplo) linhas nulas em vendas e precisasse substituir pela m√©dia:
# df["Vendas"].fillna(df["Vendas"].mean(), inplace= True) # O par√¢metro inplace = True substitui o valor em mem√≥ria
# df.dropna() # Apaga os valores nulos em todo o DataFrame
# df.dropna(subset=["Vendas"], inplace = True) # Deleta apenas se existir Null na coluna "Vendas"
```

```python
# Criando nova coluna

df["Receita"] = df["Vendas"] * df["Qtde"]
df.sample(10)
```

```python
# Retornando a maior e a menor receita
print(df["Receita"].max())
print(df["Receita"].min())
```

```python
# Retornando o top 3 registros com base em Receita
df.nlargest(3, "Receita") # As tres menores seriam df.nsmallest(3, "Receita")
```

```python
# Retornando a receita gerada por cada loja e ordenando da maior para menor
ranking = df.groupby("Cidade").sum("Receita").sort_values("Receita", ascending=False)
ranking
```

## Trabalhando com datas

```python
#Transformando tipo da coluna data em int
df["Data"] = df["Data"].astype("int64")
df.dtypes
```

```python
# Transformando o tipo da coluna data em datetime
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# Criando a coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
df.sample(5)
```

```python
# Extraindo M√™s e Dia
df["mesVenda"], df["diaVenda"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample(5)
```

```python
# Retornando o registro mais antido de acordo com a data
df["Data"].min()
```

```python
# Retornando a diferen√ßa de dias entre o primeiro registro e o da linha
df["diferencaDias"] = df["Data"] - df["Data"].min()
df.sample(5)
```

```python
# Retornando o trimestre do  registro
df["trimestreVenda"] = df["Data"].dt.quarter
df.sample(5)
```

```python
# Filtrando as vendas da loja de Recife que aconteceram em mar√ßo de 2018:
vendas_marco_19 = df.loc[(df["Cidade"] == "Recife") & (df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_marco_19
```

## Visualizando os dados

```python
# Contando os registros de LojaId
df["LojaID"].value_counts(ascending = False)
```

```python
# Gr√°fico de barras com a contagem de registros de LojaId
df["LojaID"].value_counts(ascending = False).plot.bar()
```

```python
# Gr√°fico de barras horizontais com a contagem de registros de LojaId
df["LojaID"].value_counts(ascending = True).plot.barh();
```

```python
# Gr√°fico de pizza com agrupamento de receita por ano
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie();
```

```python
# Total de venda por Cidade
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title = "Total de venda por Cidade")
plt.xlabel("Cidades")
plt.ylabel("Total de vendas")
```

```python
# Alterando a cor

df["Cidade"].value_counts().plot.bar(title = "Total de venda por Cidade", color = "g")
plt.xlabel("Cidades")
plt.ylabel("Total de vendas")
```

```python
# Alterando o estilo
plt.style.use("ggplot")

df.groupby(df["mesVenda"])["Qtde"].sum().plot(title = "Produtos vendidos por m√™s")
plt.xlabel("M√™s")
plt.ylabel("Total de vendas")
plt.legend()
```

```python
# Plotando apenas as vendas de 2019
vendas_2019 = df[df["Ano_Venda"] == 2019]

vendas_2019.groupby(vendas_2019["mesVenda"])["Qtde"].sum().plot(title = "Produtos vendidos por m√™s", marker = "o")
plt.xlabel("M√™s")
plt.ylabel("Total de vendas")
plt.legend();
```

```python
# Histograma com a quantidade de vendas de 2019
plt.hist(df["Qtde"], color="forestgreen");
```

```python
# Gr√°fico Scatter Plot
plt.scatter(vendas_2019['diaVenda'], vendas_2019["Receita"], color= "black")
plt.title("Dispers√£o de vendas/dia do m√™s para 2019")
plt.xlabel("Dia da Venda")
plt.ylabel("Valor vendido");
```

```python
# Salvando um gr√°fico em PNG

vendas_2019.groupby(vendas_2019["mesVenda"])["Qtde"].sum().plot(title = "Produtos vendidos por m√™s", marker = "o")
plt.xlabel("M√™s")
plt.ylabel("Total de vendas")
plt.legend();
plt.savefig("grafico_Produtos_vendidos_por_mes.png")
```
