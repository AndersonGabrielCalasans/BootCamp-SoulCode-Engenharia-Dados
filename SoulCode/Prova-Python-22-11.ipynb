```python
# ANA, RADAR, socorram-me subi no ônibus em Marrocos
# A  N  A       R  A  D  A  R
# 0  1  2       0  1  2  3  4
#-3 -2 -1      -5 -4 -3 -2 -1

palavra = input('''Digite a palavra e descubra se é um palíndromo
Lembre-se de não digitar palavra acentuada ou caracter especial: ''')

def Palindromo(string):
  string_min = string.lower().replace(" ", "")
  string_invert = string[::-1].lower().replace(" ", "")
  if string_min == string_invert:
    return f'A frase digitada "{palavra}" é um palíndromo'

Palindromo(palavra)
```

```python
# Crie um algoritmo em python que leia um número inteiro positivo N 
# e preencha uma lista com os N primeiros números da sequência abaixo. 
# O primeiro número da sequência é o 2. Após percorra a lista e imprima os números na tela. (2 pontos)

# 2      3        5        9        17        33        65     129       257

num = int(input('Digite um número inteiro: '))
while num < 1:
  print('Número inválido!')
  num = int(input('Digite um número inteiro positivo: '))

lista_verifica = [2, 3, 5, 9, 17, 33, 65, 129, 257]
lista = []
for i in range(num+1):
  if i in lista_verifica:
    lista.append(i)

print(f'Números da lista até o {num}:', end = ' ')
for numero in lista:
  print(numero, end = ' ')
```

```python
num = int(input('Digite quantos números terá a sequencia: '))
while num < 1:
  print('Número inválido!')
  num = int(input('Digite um número inteiro positivo: '))

lista = []
n = 2
for i in range(num):
  lista.append(n)
  n *= 2
  n -= 1

for numero in lista:
  print(numero, end = ' ')
```
