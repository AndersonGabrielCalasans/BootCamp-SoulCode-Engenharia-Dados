### Concertando inconsistencias na coluna fornecedor

#### Tipo C - MATERIAIS E SERVIÇOS DE MANUT E CONSERV DE VEÍCULOS ; PEDÁGIOS

```python
# criando filtro com 'C - MATERIAIS E SERVIÇOS DE MANUT E CONSERV DE VEÍCULOS ; PEDÁGIOS'
filtro = df.tipo == 'C - MATERIAIS E SERVIÇOS DE MANUT E CONSERV DE VEÍCULOS ; PEDÁGIOS'
df_c = df.loc[filtro]
df_c
```

##### Corrigindo PEDAGIO

```python
# Segundo filtro por palavras com PED e PEG
filtro2 = df_c.fornecedor.str.contains('PED|PEG')
df_pedagio = df_c.loc[filtro2]
df_pedagio['fornecedor'].unique()
```

```python
# Corrigindo tudo para pedágio
df.replace(['PEDÁGIO', 'CUPONS DE PEDÁGIO', 'PEDAGIO DE ELISIARIO', 'PEDAGIO',
       'PEGAGIO', 'PRAÇA DE PEDÁGIO DE GUARUJÁ', 'PEDÁGIOS'], 'PEDAGIO', inplace=True)

df
```
